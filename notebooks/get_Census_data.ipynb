```python
# ACS variables to get

acs_vars={'B01003_001E':'total population',
 'B02001_002E':'total: White',
 'B02001_003E':'total: Black',
 'B02001_005E':'total: Asian',
 'B02001_001E':'[denominator: total population for Race]',
 'B03002_012E':'total: Hispanic or Latino (any race)',
 'B03002_001E':'[denominator: total population for Hispanic or Latino]',
 'B17001_002E':'total: income below poverty level',
 'B17001_001E':'[denominator: total population for poverty]',
 'B25064_001E':'median gross rent',
 'B25070_010E': 'total: paying >50 percent of income on rent',
 'B25070_001E':'[denominator: total population for income on rent]',
 'B06011_001E':'median income'
}
```

```python
# to compute proportions

def compute_columns(df_input):

  df = df_input.copy()
  
  df['percent: White'] = df['total: White'] / df['[denominator: total population for Race]']
  df['percent: Black'] = df['total: Black'] / df['[denominator: total population for Race]']
  df['percent: Asian'] = df['total: Asian'] / df['[denominator: total population for Race]']
  df['percent: Hispanic or Latino'] = df['total: Hispanic or Latino (any race)'] / df['[denominator: total population for Hispanic or Latino]']
  df['percent: poverty'] = df['total: income below poverty level'] / df['[denominator: total population for poverty]']
  df['percent: rent burdened'] = df['total: paying >50 percent of income on rent'] / df['[denominator: total population for income on rent]']

  return df


```

```python
# get ACS data for each year

acs_year_list = []

years = tqdm(range(2019,2012, -1))

for year in years:
  
  # establish connection to api
  acs = products.ACS(year)

  # get data
  acs_data = acs.from_place('New York city, NY', 
                                level='tract', 
                                variables=list(acs_vars.keys()),
                                return_geometry=False)
  
  # rename and compute columns
  acs_data = acs_data.rename(columns=acs_vars)
  acs_data = compute_columns(acs_data)
  acs_data = acs_data.set_index('GEOID')
  acs_data['year'] = year

  # prepend year
  #acs_data = pd.concat({year: acs_data}, names=['Year'])

  acs_year_list.append(acs_data)

acs_year = pd.concat(acs_year_list)
```
